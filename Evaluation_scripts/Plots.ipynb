```python
import matplotlib.pyplot as plt 

fig = plt.figure(figsize=(10, 7))
host = fig.add_subplot(111)

par1 = host.twinx()

host.set_ylim(0.06, 0.25)
par1.set_ylim(0.30, 0.45)

host.set_xlabel("Imbalance Type", fontsize = 18)
host.set_ylabel("SSIM CIFAR10", fontsize = 18)
par1.set_ylabel("SSIM MNIST", fontsize = 18)


p1, = host.plot(['Linear\nImbalance','μ = 2','μ = 5','μ = 7','μ = 9'], [0.082, 0.0715, 0.0650, 0.097, 0.20], color = 'darkblue',
                marker = 'o', linestyle = '-', label = "FocalGAN_Cifar10")
p2, = host.plot(['Linear\nImbalance','μ = 2','μ = 5','μ = 7','μ = 9'], [0.118, 0.0875, 0.0786, 0.133, 0.212], color = 'cornflowerblue',
                marker = 'p', linestyle = '-.', label = "ACGAN_Cifar10")
p3, = par1.plot(['Linear\nImbalance','μ = 2','μ = 5','μ = 7','μ = 9'], [0.37, 0.41, 0.36, 0.37, 0.38], color = 'maroon',
                marker = 'o', linestyle = '-', label = "FocalGAN_MNIST")
p4, = par1.plot(['Linear\nImbalance','μ = 2','μ = 5','μ = 7','μ = 9'], [0.37, 0.42, 0.35, 0.37, 0.40], color = 'lightcoral',
                marker = 'p', linestyle = '-.', label = "ACGAN_MNIST")

lns = [p3, p4, p1, p2]
host.legend(handles=lns, loc='best', prop={'size': 14})
for tick in host.xaxis.get_major_ticks():
    tick.label.set_fontsize(14)
for tick in host.yaxis.get_major_ticks():
    tick.label.set_fontsize(14)
for tick in par1.yaxis.get_major_ticks():
    tick.label.set_fontsize(14)
plt.title('Mean-SSIM comparison', fontsize = 22)
```
